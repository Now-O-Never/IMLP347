```python
from sklearn.ensemble import BaggingClassifier
from sklearn import datasets
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris=datasets.load_iris()
X=iris.data
y=iris.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.3,random_state=0)
```

```python
len(X_test)
```

```python
from sklearn import tree
from sklearn import svm
from sklearn.naive_bayes import GaussianNB
```

```python
clf=tree.DecisionTreeClassifier()
# clf=svm.LinearSVC()
# clf=GaussianNB()
```

```python
bagging=BaggingClassifier(base_estimator=clf,n_estimators=10, # n_estimator = 要生成幾個分類器
                          bootstrap=True,bootstrap_features=True,max_features=3,max_samples=0.7)
                          #bootstrap = True 表示每次訓練會把特徵值取後放回 
```

```python
bagging.fit(X_train,y_train)
```

```python
bagging.predict(X_test)
```

```python
bagging.score(X_train,y_train)
```

```python
bagging.score(X_test,y_test)
```
